```python
# Task 8.

class DataSet:
    def __init__(self, filename, threshold=75):
        self.filename = filename
        self.data = []
        self.total = 0
        self.count = 0
        self.average = 0
        self.minimum = None
        self.maximum = None
        self.threshold = threshold

    # Load data from file
    def load_data(self):
        try:
            file = open(self.filename, "r")
            lines = file.readlines()
            file.close()

            if len(lines) == 0:
                print("Error: The file is empty.")
                return False

            for line in lines:
                line = line.strip()
                if line.isdigit():
                    self.data.append(int(line))
                else:
                    print("Invalid value skipped:", line)

            if len(self.data) == 0:
                print("Error: No valid numerical data found.")
                return False

            return True

        except FileNotFoundError:
            print("Error: File not found.")
            return False

    # Calculate statistics
    def calculate_statistics(self):
        if len(self.data) == 0:
            return

        self.total = 0
        self.minimum = self.data[0]
        self.maximum = self.data[0]
        self.count = 0

        for value in self.data:
            self.total += value
            self.count += 1
            if value < self.minimum:
                self.minimum = value
            if value > self.maximum:
                self.maximum = value

        self.average = self.total / self.count

    # Display results
    def display_results(self):
        if len(self.data) == 0:
            return

        print("\n--- Analysis Results ---")
        print("Total:", self.total)
        print("Number of data points:", self.count)
        print("Average:", self.average)
        print("Minimum:", self.minimum)
        print("Maximum:", self.maximum)

        if self.average > self.threshold:
            print("Performance: High Performance")
        else:
            print("Performance: Needs Improvement")

    # Save results to a report file
    def save_results(self, report_filename):
        try:
            with open(report_filename, "w") as file:
                file.write("--- Analysis Report ---\n")
                file.write(f"Total: {self.total}\n")
                file.write(f"Number of data points: {self.count}\n")
                file.write(f"Average: {self.average}\n")
                file.write(f"Minimum: {self.minimum}\n")
                file.write(f"Maximum: {self.maximum}\n")
                if self.average > self.threshold:
                    file.write("Performance: High Performance\n")
                else:
                    file.write("Performance: Needs Improvement\n")
            print(f"\nResults saved to {report_filename}")

        except Exception as e:
            print("Error saving report:", e)


# --- Using the class ---
dataset = DataSet("marks.txt", threshold=75)

if dataset.load_data():
    dataset.calculate_statistics()
    dataset.display_results()
    dataset.save_results("report.txt")
```
